```python
class Node:
    def __init__(self, name, mark):
        self.name = name
        self.mark = mark
        self.right = None
        self.left = None
        self.parent = None
    
    def __repr__(self):
        return self.name
```

```python
class Manager:
    def __init__(self, root):
        self.root = root
    
    def add(self, new_node):
        cursore = self.root
        while True:
            if new_node.mark > cursore.mark:
                if cursore.right is not None:
                    cursore = cursore.right
                else:
                    cursore.right = new_node
                    new_node.parent = cursore
                    break
            else:
                if cursore.left is not None:
                    cursore = cursore.left
                else:
                    cursore.left = new_node
                    new_node.parent = cursore
                    break
                    
    def find(self, mark):
        cursore = self.root
        while True:
            if mark == cursore.mark:
                return cursore
            if mark > cursore.mark:
                if cursore.right is not None:
                    cursore = cursore.right
                else:
                    return None
            else:
                if cursore.left is not None:
                    cursore = cursore.left
                else:
                    return None
                
    def find_all(self, mark):
        cursore = self.root
        result = []
        while True:
            if mark == cursore.mark:
                result.append(cursore.name)
            if mark > cursore.mark:
                if cursore.right is not None:
                    cursore = cursore.right
                else:
                    return result
            else:
                if cursore.left is not None:
                    cursore = cursore.left
                else:
                    return result
    # new           
    def get_min(self, vertex): 
        while vertex.left != None:
            vertex = vertex.left
        return vertex
    # new
    def delete(self, mark):
        if not self.find(mark):
            return 'Вершина не в дереве'
        
        parent = None
        cursore = self.root
        replace = None
        while(cursore != None and cursore.mark != mark):
            parent = cursore
            if mark >= cursore.mark:
                cursore = cursore.right
                flag = True
            else:
                cursore = cursore.left
                flag = False
    
        if cursore.left == None and cursore.right == None:
            if flag:
                parent.right = None
            else:
                parent.left = None
            del cursore
            return 'Удалено!'
        
        elif cursore.left == None:
            if flag:
                parent.right = cursore.right
            else:
                parent.left = cursore.left
            del cursore
            return 'УДалено'
        
        elif cursore.right == None:
            if flag:
                parent.left = cursore.left
            else:
                parent.right = cursore.right
            del cursore
            return 'Удалено0!'
        
        else:
            replace = self.get_min(cursore.right)
            tmp = replace.mark
            self.delete(replace.mark)
            cursore.mark = tmp
```

```python
root = Node("Egor", 20)
m = Manager(root)
```
